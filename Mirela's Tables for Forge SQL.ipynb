```python
Guests = pd.read_csv('/Users/mirelalynch/Desktop/Project 2 Launch  - Guests.csv')
Rooms = pd.read_csv('/Users/mirelalynch/Desktop/Project 2 Launch  - Rooms.csv')
Reservations = pd.read_csv('/Users/mirelalynch/Desktop/Project 2 Launch  - Reservations.csv')
Payments = pd.read_csv('/Users/mirelalynch/Desktop/Project 2 Launch  - Payments.csv')
```

```python
table_name1 = 'Guests'
table_name2 = 'Rooms'
table_name3 = 'Reservations'
table_name4 = 'Payments'
```

```python
Guests.columns = Guests.columns.str.strip()
Rooms.columns = Rooms.columns.str.strip()
Reservations.columns = Reservations.columns.str.strip()
Payments.columns = Payments.columns.str.strip()
```

```python
Guests.to_sql(name=table_name1, con=connection, if_exists='replace', index=False)
Rooms.to_sql(name=table_name2, con=connection, if_exists='replace', index=False)
Reservations.to_sql(name=table_name3, con=connection, if_exists='replace', index=False)
Payments.to_sql(name=table_name4, con=connection, if_exists='replace', index=False)
```

```python
query1 = "SELECT * FROM Guests"
query2 = "SELECT * FROM Rooms"
query3 = "SELECT * FROM Reservations"
query4 = "SELECT * FROM Payments"
```
